### *Visualization of Sales,Quantity,Discount and Profit using histogram*

```
data.hist(figsize=(10,10),bins=50)
plt.show()
```

```
pd.DataFrame(data.groupby('State').sum())['Profit'].sort_values(ascending=True)
```

```
sns.lineplot(x="Discount", y="Profit",label = "Profit",data=data)
plt.legend()
plt.show()
```

### *Visualizing profit and sales across different segments*

```
data.groupby("Segment")[['Profit', 'Sales']].sum().plot.bar(color = ["blue", "green"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

### *Visualizing segment wise sales distribution in each region*

```
plt.figure(figsize=(12,8))
plt.title("Distribution of Segment wise sales in each region " , fontsize=15)
sns.barplot(x="Region",y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='Blues')
plt.xlabel('Region', fontsize=15)
plt.tick_params(labelsize=12) 
plt.show()
```

### *Visualizing Profit and sales across different regions*

```
data.groupby("Region")[['Profit', 'Sales']].sum().plot.bar(color = ["green", "yellow"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

### *Visualizing profit and sales across states*

```
dt = data.groupby("State")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False)
dt[:].plot.bar(color = ["Red", "blue"],figsize= (18,8))
plt.title("Profit/Loss & Sales across States")
plt.xlabel("States")
plt.ylabel("Profit/Loss & Sales")
plt.show()
```

### *Visualizing Profit and Sales across different Categories*

```
data.groupby("Category")[['Profit', 'Sales']].sum().plot.bar(color = ["green", "yellow"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

### *Visualizing profit and sales across various Sub-Categories*

```
sc=data.groupby("Sub-Category")[["Sales","Profit"]].mean().sort_values(by = "Sales", ascending = False)
sc.plot.bar(color=['red','blue'],figsize= (18,8))
plt.title("Profit and Sales across various Sub-Categories",fontsize=20)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel("Profit/Loss & Sales")
plt.tick_params(labelsize=14)
plt.show()
```
